```python
qsts = dss_qsts.DSS_Timeseries(
    redirects=cktfile,
    time_step="60m",
    simulation_steps=24*365,
    simulation_mode="duty",
    simulation_controlmode='static',
    precompile=False
)
qsts.compile_dss()
qsts.run()
node_names = qsts.dss.Circuit.AllNodeNames()
Vm2 = (qsts.vmags_pu_mvts**2)
P = np.real(qsts.complex_powers_mvts)
Q = np.imag(qsts.complex_powers_mvts)
```

```python
# nodes to study
node_names = qsts.dss.Circuit.AllNodeNames()
study_node_names = node_names[:3]
study_nodes = np.where(np.isin(node_names,study_node_names))[0]
print('study_nodes:',study_nodes)
print('study_node_names',study_node_names)
```

## Fit three-phase CLAs

### Fit an underestimating CLA for every node in the network

```python
underest = yadi_cla.CLA(verbose=False,maxiters=5000)
n = np.shape(Vm2)[1]
dvp_under,dvq_under,v0_under = [],[],[]
for i in tqdm(study_nodes):
    _dvp,_dvq,_v0 = underest.fit_pq(P,Q,Vm2[:,i],ub=Vm2[:,i])
    dvp_under.append(_dvp)
    dvq_under.append(_dvq)
    v0_under.append(_v0)
```

### Fit an Overestimating CLA for every node in the network

```python
overest = yadi_cla.CLA(verbose=False,maxiters=1000)
n = np.shape(Vm2)[1]
dvp_over,dvq_over,v0_over = [],[],[]
for i in tqdm(study_nodes):
    _dvp,_dvq,_v0 = overest.fit_pq(P,Q,Vm2[:,i],lb=Vm2[:,i])
    dvp_over.append(_dvp)
    dvq_over.append(_dvq)
    v0_over.append(_v0)
```

```python
itest = 0
nodetest = study_nodes[itest]
(m,n) = P.shape
vhat_over = v0_over[itest][0] + P@dvp_over[itest] + Q@dvq_over[itest]
vhat_under = v0_under[itest][0] + P@dvp_under[itest] + Q@dvq_under[itest]
# print the relative errors
print('Relative error in voltage magnitude at node',study_node_names[itest])
print('Overestimation:',np.linalg.norm(vhat_over-Vm2[:,nodetest])/np.linalg.norm(Vm2[:,nodetest]))
```

```python

plt.scatter(x=P[:,nodetest],y=Vm2[:,nodetest])
plt.scatter(x=P[:,nodetest],y=vhat_under,color='red',label='CLA')
plt.scatter(x=P[:,nodetest],y=vhat_over,color='green',label='CLA')

plt.xlabel("Active power (kW)")
plt.ylabel("Voltage magnitude squared (pu)")
```

## CLA effectiveness by Phase

Plot the overestimating/underestimating performance of the CLA by phase in 3 subfigures.

```python
import pandas as pd
m = len(Vm2[:,0])
df = pd.DataFrame(columns=['vm2_gt','vm2_hat','phase','type'])
for experiment in [['overestimate',dvp_over,dvq_over,v0_over],['underestimate',dvp_under,dvq_under,v0_under]]:
    (type,dvp,dvq,v0) = experiment
    for (phase,dvp_ph,dvq_ph,v0_ph,node) in zip(['a','b','c'],dvp,dvq,v0,study_nodes):
        vhat = v0_ph[0] + P@dvp_ph + Q@dvq_ph
        df_ph = pd.DataFrame({
            'vm2_gt':Vm2[:,node],
            'vm2_hat':vhat,
            'phase':[phase]*m,
            'type':[type]*m
        })
        df = pd.concat([df,df_ph])

g= sns.relplot(
    df,
    x='vm2_gt',
    y='vm2_hat',
    hue='type',
    col='phase')
# g.map(plt.plot([1.02,1.15],[1.02,1.15],color='black',linestyle='--'))
```

```python

def get_nodes_by_phase(node_names=node_names):
    """
    Returns a dictionary whose keys are phases a b or c and which 
    contains the indeces of all nodes that have that phase
    """
    phase_label = {
        'a':'.1',
        'b':'.2',
        'c':'.3'
        }
    nodes_by_phase = {}
    for phase in ['a','b','c']:
        label = phase_label[phase]
        nodes_by_phase[phase] = {
            'idx':[],
            'names':[]
        }
        for i,name in enumerate(node_names):
            if label in name:
                nodes_by_phase[phase]['names'].append(name)
                nodes_by_phase[phase]['idx'].append(i)

    return nodes_by_phase

nodes_by_phase = get_nodes_by_phase()
```
